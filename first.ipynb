```python
import matplotlib.pyplot as plt
import numpy as np
x=np.linspace(-np.pi,np.pi,256,endpoint=True)
c,s=np.cos(x),np.sin(x)
plt.figure(1)


plt.plot(x,c,color="red")
plt.plot(x,s,color="blue")
plt.title("cos,sin")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
a=7
b=6
fig=plt.figure(figsize=(10,10))
ax=fig.add_subplot(111)
theta=np.arange(0,2*np.pi,np.pi/100)
x=(a-2)*np.cos(theta)
y=(b-2)*np.sin(theta)
x1=a*np.cos(theta)
y1=b*np.sin(theta)



x2=2*np.cos(theta*5)
y2=2*np.sin(theta*5)
ax.plot(x,y)
ax.plot(x1,y1)
ax.plot(x1+x2,y1+y2)
plt.show()
plt.clf()
```
